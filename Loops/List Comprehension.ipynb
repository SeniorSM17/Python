```python
#Get the index and the value as a tuple for items in the list “hi”, 4, 8.99, ‘apple’, (‘t,b’,’n’). 
#Result would look like (index, value), (index, value) 
tp=("hi", 4, 8.99, 'apple', ('t','b','n'))
print([(tp.index(i),i) for i in tp])
```

```python
# Find the common numbers in two lists (without using a tuple or set) list_a = 1, 2, 3, 4, list_b = 2, 3, 4, 5 
list_a = [1, 2, 3, 4, ]
list_b = [2, 3, 4, 5 ]
print([i for i in list_a for j in list_b if i==j])
```

```python
#Given numbers = range(20), produce a list containing the word ‘even’ if a number in the numbers is even, and the word ‘odd’ 
#if the number is odd. Result would look like ‘odd’,’odd’, ‘even’ 
print(["even" if i%2==0 else "odd" for i in range(21)])
```

```python
#Produce a list of tuples consisting of only the matching numbers in these lists list_a = 1, 2, 3,4,5,6,7,8,9, list_b = 2, 7, 1, 12.
#Result would look like (4,4), (12,12) 
list_a = [1, 2, 3, 4, ]
list_b = [2, 3, 4, 5 ]
print([(i,j) for i in list_a for j in list_b if i==j])
```

```python
# Find all of the words in a string that are less than 4 letters 
print([i for i in input("Enter the string ").split() if len(i) <=4])
```

```python
# Use a nested list comprehension to find all of the numbers from 1-1000 that are divisible by any single digit besides 1 (2-9) 
print([i for i in range(1,1000) for j in [2,3,4,5,6,7,8,9] if i%j==0])
```
